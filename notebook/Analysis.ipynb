```python
import json

world_cup_years = [
    1975, 1979, 1983, 1987,
    1992, 1996, 1999, 2003,
    2007, 2011, 2015, 2019, 2023
]

india_wc_matches = []

for file in files:

    if not file.endswith(".json"):
        continue

    path = "../data/raw/odis/all_matches/" + file

    try:
        with open(path, "r", encoding="utf-8") as f:
            match = json.load(f)

        info = match["info"]

        teams = info["teams"]

        date = info["dates"][0]

        year = int(date[:4])

        event = info.get("event", {}).get("name", "").lower()

        if (
            "India" in teams
            and year in world_cup_years
            and "world cup" in event
        ):
            india_wc_matches.append(match)

    except:
        pass

print("India World Cup Matches:", len(india_wc_matches))
```

```python
import json

india_matches = []

for file in files:

    if not file.endswith(".json"):
        continue

    path = "../data/raw/odis/all_matches/" + file

    try:
        with open(path, "r", encoding="utf-8") as f:
            match = json.load(f)

        teams = match["info"]["teams"]

        # Only India matches
        if "India" in teams:
            india_matches.append(match)

    except:
        pass

print("Total India Matches:", len(india_matches)) 

```

```python
world_cup_editions = set()

for match in india_matches:

    try:
        info = match["info"]

        event = info.get("event", {}).get("name", "")

        if "world cup" in event.lower():

            year = info["dates"][0][:4]

            world_cup_editions.add(year)

    except:
        pass

print("World Cup Editions Found:")
print(sorted(world_cup_editions))

print("Total Editions:", len(world_cup_editions))
```

```python
wc_matches_by_edition = {}

for match in india_matches:

    try:
        info = match["info"]

        event = info.get("event", {}).get("name", "")

        if "world cup" in event.lower():

            year = info["dates"][0][:4]

            if year not in wc_matches_by_edition:
                wc_matches_by_edition[year] = 0

            wc_matches_by_edition[year] += 1

    except:
        pass

print("India Matches Per World Cup Edition:\n")

for year, matches in sorted(wc_matches_by_edition.items()):
    print(year, ":", matches)
```

```python
wc_matches_by_edition = {}

for match in india_wc_matches:

    try:
        info = match["info"]

        year = info["dates"][0][:4]

        if year not in wc_matches_by_edition:
            wc_matches_by_edition[year] = 0

        wc_matches_by_edition[year] += 1

    except:
        pass

print("India Matches Per World Cup Edition:\n")

for year, matches in sorted(wc_matches_by_edition.items()):
    print(year, ":", matches)
```

```python
import pandas as pd

matches_data = []

for match in india_wc_matches:

    try:
        info = match["info"]

        matches_data.append({
            "date": info["dates"][0],
            "team1": info["teams"][0],
            "team2": info["teams"][1],
            "venue": info.get("venue"),
            "city": info.get("city"),
            "event": info.get("event", {}).get("name", "")
        })

    except:
        pass

df = pd.DataFrame(matches_data)

df.head()
```
